```python
import yaml
from shasta.visualize import plot_nodes
import osmnx as ox
from matplotlib import pyplot as plt
```

```python
config_path2 = 'config/test_config.yml'
config2 = yaml.load(open(str(config_path2)), Loader=yaml.SafeLoader)
```

```python
plot_nodes(config=config2)
```

```python
read_path = '/'.join([
        config2['urdf_data_path'], config2['simulation']['map_to_use'], 'map.osm'
    ])
```

```python
G = ox.graph_from_xml(read_path, simplify=False, bidirectional='walk')
G = ox.simplification.simplify_graph(G)
```

```python
ox.plot_graph(G,figsize=(20,30))
```

```python
edges = ox.graph_to_gdfs(G, nodes=False, edges=True)
edges_series = edges['length'] #gives you a pandas series with edge lengths
edges[['osmid','length']]
```
